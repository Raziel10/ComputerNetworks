```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0, 5.0, 0.01)
s = t / (1+t)
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot()
ax.plot(t, s)

ax.set(xlabel='G', ylabel='S = G / (1+G)',
       title='CSMA NP caudal')
ax.grid()

fig.savefig("CSMA_NP_throughput_2D.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 5.1, 0.1)
y = np.arange(0, 5.1, 0.1)

# Compute z to make the pringle surface.
x, y = np.meshgrid(x, y) 
z = (x * y / 0.01 * np.exp((-x)* y)) / ((x * y / 0.01 ) * (1 + (2 * 0.01)) + np.exp((-x)* y))


fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection='3d')
ax.plot_wireframe(x, y, z, rstride=10, cstride=10)
ax.set_title("3D CSMA NP caudal ",fontsize=14,fontweight="bold")
ax.set_xlabel("lambda")
ax.set_ylabel("T")
ax.set_zlabel("S")
fig.savefig("CSMA_NP_throughput_3D.png")
plt.show()
```

```python

from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
 
 
# Creating dataset

x = np.arange(0, 5.0, 0.1)
y = np.arange(0, 5.0, 0.1)
x, y = np.meshgrid(x, y)
z = (x * y / 0.01 * np.exp((-x)* y)) / ((x * y / 0.01 ) * (1 + (2 * 0.01)) + np.exp((-x)* y))
 
# Creating figure
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection='3d')
ax.set_title("3D CSMA NP caudal ",fontsize=14,fontweight="bold")
ax.set_xlabel("lambda")
ax.set_ylabel("T")
ax.set_zlabel("S")
 
# Creating plot
ax.contour3D(x, y, z, 50, cmap='binary')
fig.savefig("CSMA_NP_throughput.png")
# show plot
plt.show()
```

```python
f = open("Data/results.txt", "r")
x_exp = []
y_exp = []
z_exp = []
for line in f:
    values = line.split(",")
    #print(values)
    if( values[0] == 'l'):
        continue;
    x_exp.append(float(values[0]))
    y_exp.append(float(values[1]))
    z_exp.append(float(values[2]))

x_exp = np.array(x_exp)
y_exp = np.array(y_exp)
z_exp = np.array(z_exp)
#transform values into shape
x = x_exp.reshape([50, 50])
y = y_exp.reshape([50, 50])
z = z_exp.reshape([50, 50])

#y , z = np.meshgrid(y_exp, z_exp)


fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection ='3d')
ax.set_title("Grafica 3D Aloha Simulado caudal ",fontsize=14,fontweight="bold")
ax.set_xlabel("lambda")
ax.set_ylabel("T")
ax.set_zlabel("S")
 
# Creating plot
ax.contour3D(x, y, z, 50, cmap='binary')
 
# show plot
plt.show()

fig.savefig("Aloha_Simulation_throughput.png")
```

```python


# Compute z to make aloha surface.
z_aloha = x * y * np.exp((-2) * x * y)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection='3d')
ax.plot_wireframe(x, y, z_aloha, rstride=10, cstride=10)
ax.plot_wireframe(x, y, z, rstride=10, cstride=10, color='r')
ax.set_title("Grafica 3D Aloha teorico y simulado caudal ",fontsize=14,fontweight="bold")
ax.set_xlabel("lambda")
ax.set_ylabel("T")
ax.set_zlabel("S")

plt.show()
fig.savefig("Aloha_and_Simulation_throughput.png")
```
